```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\nilvi\OneDrive\Escritorio\Ironhack\Week4\Group Project\football\top5scorers.csv")
df
```

```python
df['surname'] = df['playerName'].apply(lambda name: name.split()[-1])
```

```python
df
```

```python
from adjustText import adjust_text

plt.figure(figsize=(14, 8))
sns.lineplot(data=df, x="season", y="goals_scored", hue="surname", marker="o", palette="tab10")

# Prepare text objects
texts = []
for i, row in df.iterrows():
    label = f"{row['surname']}\n{round(row['goal_conversion_percent'], 1)}%"
    texts.append(
        plt.text(row['season'], row['goals_scored'], label, fontsize=8, ha='center')
    )

# Adjust to avoid overlaps
adjust_text(texts, arrowprops=dict(arrowstyle='-', color='gray'))

# Final plot touches
plt.title("Top 5 Scorers per Season (2014–2020), % conversion rate (goals/shot)", fontsize=16)
plt.ylabel("Goals Scored")
plt.xlabel("Season")
plt.xticks(sorted(df['season'].unique()))
plt.legend().remove()
plt.tight_layout()
plt.show()
```
