```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = r'C:\Users\rergu\OneDrive\IronHack\MiniProjects\football\football\ft_aggregated.csv'
df = pd.read_csv(url, index_col=None)
```

```python
df.sample(10)
```

```python
df.columns
```

```python
# Calculate the correlation coefficient XG and Goals
correlation = df['xG_per_game'].corr(df['GF_per_game'])
print(f"Correlation between xG_per_game and GF_per_game: {correlation:.2f}")

# Visualize with a scatterplot and regression line
plt.figure(figsize=(4, 3))
sns.regplot(x='xG_per_game', y='GF_per_game', data=df, scatter_kws={"s": 20}, line_kws={"color": "red"})
plt.title('Correlation between xG per Game and Goals For per Game')
plt.xlabel('xG per Game')
plt.ylabel('Goals For per Game')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation coefficient Shots and Goals

correlation = df['shots_per_game'].corr(df['GF_per_game'])
print(f"Correlation between Shots on Target and Goals per game: {correlation:.2f}")

plt.figure(figsize=(4, 3))
sns.regplot(x='GF_per_game', y='shots_per_game', data=df, scatter_kws={"s":20}, line_kws={"color": "red"})
plt.title(f"Correlation between Shots and Goals per game: {correlation:.2f}")
plt.xlabel('Goals per game')
plt.ylabel('Shots per game')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation coefficient Shots on Target and Points

correlation = df['shots_per_game'].corr(df['points_per_game'])
print(f"Correlation between Shots on Target and Points per game: {correlation:.2f}")

plt.figure(figsize=(4, 3))
sns.regplot(x= 'points_per_game', y='shots_per_game', data=df, scatter_kws={'s':20}, line_kws={'color':'red'})
plt.title(f'Correlation between Shots and Points Per Game: {correlation:.2f}')
plt.xlabel('Points Per Game')
plt.ylabel('Shots on Target Per Game')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation coefficient xG and Points
correlation = df['GF_per_game'].corr(df['points_per_game'])
print(f'Correlation between xG and Points Per Game: {correlation:.2f}')

plt.figure(figsize=(4, 3))
sns.regplot(x= 'points_per_game', y='GF_per_game', data=df, scatter_kws={'s':20}, line_kws={'color':'red'})
plt.title(f'Correlation between Goals and Points Per Game: {correlation:.2f}')
plt.xlabel('Point Per Game')
plt.ylabel('Goals Per Game')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation coefficient corner and goals
correlation = df['corners_per_game'].corr(df['GF_per_game'])
print(f'Correlation between Corners and Goals per Game: {correlation:.2f}')

plt.figure(figsize=(4, 3))
sns.regplot(x='corners_per_game', y='GF_per_game', data=df, scatter_kws={'s':20}, line_kws={'color':'red'})
plt.title(f'Correlation between Corners and Goals per Game: {correlation:.2f}')
plt.xlabel('Goals Per Game')
plt.ylabel('Corners Per Game')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation coefficient corner and points
correlation = df['corners_per_game'].corr(df['points_per_game'])
print(f'Correlation between Corners and Points per Game: {correlation:.2f}')

plt.figure(figsize=(4, 3))
sns.regplot(x='points_per_game', y='corners_per_game', data=df, scatter_kws={'s':20}, line_kws={'color': 'red'})
plt.title(f'Correlation between Corners and Points per Game: {correlation:.2f}')
plt.xlabel('Points Per Game')
plt.ylabel('Corners Per Game')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation coefficient corner and points
correlation = df['ht_won'].corr(df['won'])
print(f'Correlation between HT Win Ratio and Points per Game: {correlation:.2f}')

plt.figure(figsize=(4, 3))
sns.regplot(x='won', y='ht_won', data=df, scatter_kws={'s':50}, line_kws={'color': 'red'})
plt.title('Correlation between HT Win and FT Win')
plt.xlabel('FT Win')
plt.ylabel('HT Win')
plt.grid(True)
plt.tight_layout()
plt.show()
```
